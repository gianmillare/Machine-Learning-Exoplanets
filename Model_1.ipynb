```python
import pandas as pd
import numpy as np
```

# Import the Master Data File

```python
df = pd.read_csv("Resources/cumulative.csv")
df.head()
```

# Preprocessing - Data Cleaning

```python
exoplanetsdf = df.drop(columns=["rowid", "kepid", "kepler_name", "koi_teq_err1", "koi_teq_err2", "koi_tce_plnt_num",
                               "koi_tce_delivname", "kepoi_name", "koi_pdisposition"])
exoplanetsdf.head()
```

```python
exoplanetsdf = exoplanetsdf.dropna(axis='columns', how='all')
exoplanetsdf = exoplanetsdf.dropna()
exoplanetsdf.head()
```

# Begin Data Processing

```python
X = exoplanetsdf.drop("koi_disposition", axis=1)
y = exoplanetsdf["koi_disposition"]
feature_names = X.columns
print(X.shape, y.shape)
```

# Implement Train_Test_Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X.head()
```

# Scale the data with MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
```

# Transform the scaled data

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Use Support Vector Machine and fit the model

```python
from sklearn.svm import SVC 

model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

# Use model.score to value R2 Score

```python
print(f"Training Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Score: {model.score(X_test_scaled, y_test)}")
```

# Implementing GridSearch --> Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Fit the model using the grid search estimator (grid.fit)
# This will take the SVC model and try each combination of parameters

grid.fit(X_train_scaled, y_train)
```

# List the best parameters and score

```python
print(grid.best_score_)
print(grid.best_params_)
```

# Make a Prediction

```python
predictions = grid.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions))
```
